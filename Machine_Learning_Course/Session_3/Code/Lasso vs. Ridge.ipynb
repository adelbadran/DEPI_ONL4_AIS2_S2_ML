# Lasso vs. Ridge
---

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = np.array([[1], [2], [3], [4]])

y = np.array([2, 2.8, 3.6, 4.5])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_lin = lin_reg.predict(x)
print(lin_reg.coef_, lin_reg.intercept_)
```

```python
lin_ridge = Ridge(alpha=1)
lin_ridge.fit(x, y)
y_pred_ridge = lin_ridge.predict(x)
print(lin_ridge.coef_, lin_ridge.intercept_)
```

```python
lin_liasso = Lasso(alpha=0.1)
lin_liasso.fit(x, y)
y_pred_lasso = lin_liasso.predict(x)
print(lin_liasso.coef_, lin_liasso.intercept_)
```

```python
print("Linear Regression MSE:", mean_squared_error(y, y_pred_lin))
print("LinearRegression")
print("r2_score:", r2_score(y, y_pred_lin))
```

```python
print("Linear Regression MSE:", mean_squared_error(y, y_pred_lasso))
print("LinearRegression")
print("r2_score:", r2_score(y, y_pred_lasso))
```

```python
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred_lin, color='red', label='Linear Regression')
plt.plot(x, y_pred_lasso, color='green',linestyle='--', label='lasso Regression')
plt.plot(x, y_pred_ridge, color='orange',linestyle='-', label='ridge Regression')
plt.legend()
plt.xlabel('x')
plt.title('Comparison of Regression Models')
plt.ylabel('y')
```
