## Visualization using Matplotlib
---

```python
# Once Matplotlib is installed, import it in your applications by adding the import module statement:

import matplotlib.pyplot as plt # for data visualization

import numpy as np # for numerical operations
```

```python
# Plotting with line

x_points = np.array([5, 10])
y_points = np.array([50, 100])

plt.plot(x_points, y_points) # The plot() function is used to draw points (markers) in a diagram.
plt.show()
```

```python
# Plotting without line (Points)

plt.plot(x_points, y_points, 'o')
plt.show()
```

```python
# Default ??

y_points_def = [10, 40, 5, 90]
plt.plot(y_points_def) # Default x : [0, 1, 2, 3, 4]
plt.show()
```

```python
# Marker = 'o'

y_points_def = [10, 40, 5, 90]
plt.plot(y_points_def, marker='o') # try '*'
plt.show()
```

### Marker Reference
#### You can choose any of these markers:



```python
# Format Strings (fmt) as marker|line|color

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b')
plt.show()
```

### Line Reference

#### The line value can be one of the following:



### Color Reference

#### The short color value can be one of the following:



```python
# Marker Size (ms)

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
# Marker Color

# You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'k', mfc = 'r') # mec : marker edge color - mfc : marker face color 
plt.show()
```

```python
# line style

plt.plot(ypoints, linestyle = 'dotted') #short syntex : plt.plot(ypoints, ls = ':')
plt.show()
```

### Line sytyle

#### You can choose any of these styles:



```python
# Line color

plt.plot(ypoints, color = '#4CAF50') # or use : plt.plot(ypoints, c = 'g')
plt.show()
```

```python
# Line width

plt.plot(ypoints, linewidth = '20.5') # or use : (lw) instead of linewidth
plt.show()
```

```python
# Multible lines

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
# Labels and Title

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Averge Pulse") # X axis
plt.ylabel("Calorie Burnage") # Y axis
plt.title("Sports Watch Data") # Title

plt.show
```

```python
# Set Font Properties for Title and Label

font1 = {"family":"serif", "color":"blue", "size":20}
font2 = {"family":"serif", "color":"darkred", "size":15}


plt.xlabel("Averge Pulse", fontdict=font2) # X axis
plt.ylabel("Calorie Burnage", fontdict=font2) # Y axis
plt.title("Sports Watch Data", fontdict=font1, loc="left") # Title {fontdict, loc(ation)}

plt.plot(x, y)
plt.show()
```

```python
# Add Grid Lines to a Plot

plt.xlabel("Averge Pulse", fontdict=font2) # X axis
plt.ylabel("Calorie Burnage", fontdict=font2) # Y axis
plt.title("Sports Watch Data", fontdict=font1, loc="left") # Title {fontdict, loc(ation)}

plt.grid() # try : plt.grid(axis = 'x') - plt.grid(axis = 'y')
plt.plot(x, y)
plt.show()
```

```python
# Set Line Properties for the Grid

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

```python
# Display Multiple Plots (subplot)

x = [1, 2, 3, 4, 5]
y = [10, 20 , 30, 40, 50]

plt.subplot(1, 2, 1) # (Row, Colmn, index) - the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x, y)
plt.title("Sales")

x = [-1, -2, -3, -4, -5]
y = [10, 20 , 30, 40, 50]

plt.subplot(1, 2, 2) # (Row, Colmn, index) - the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("Age")
plt.plot(x, y)

plt.suptitle("My Shop")
plt.show()
```

#### The Matplotlib module has a number of available colormaps.



```python
# Create a color array, and specify a colormap in the scatter plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis') # You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
plt.colorbar() # Display ColorMap

plt.show()
```

```python
# Size

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5) # alpha = transparency (0 - 1)

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='RdYlBu')

plt.colorbar()

plt.show()
```

```python
# Histogram : is a graph showing frequency distributions.

x = np.random.normal(170, 10, 250) # (arround, std, n_values)

plt.hist(x, color="#C86A6A")
plt.show()
```

```python
# Pie charts

y = [35, 25, 25, 15]

plt.pie(y) # The size of each wedge is determined by comparing the value with all the other values, by using this formula: The value divided by the sum of all values: x/sum(x)
plt.show()
```

```python
# With labels

y = [35, 25, 25, 15]
labels = ["IS", "CS", "MI", "DS"]

plt.pie(y, labels=labels) # The size of each wedge is determined by comparing the value with all the other values, by using this formula: The value divided by the sum of all values: x/sum(x)
plt.show() # By default the plotting of the first wedge starts from the x-axis (angle = 0) and moves counterclockwise
```

```python
# Change start angle

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
# Explode : Slices

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.5, 0.1, 0.8] # Partition

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# Shadow + Colors

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True,  colors = mycolors)
plt.show()
```

```python
# Legend : To add a list of explanation for each wedge, use the legend() function

# Change start angle

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title="Four Fruits:")
plt.show()
```
