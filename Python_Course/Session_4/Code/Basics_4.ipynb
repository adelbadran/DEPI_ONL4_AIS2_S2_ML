```python
# import necessary libraries

import os
```

```python
# open a file using 'with' statement

with open('test.txt', 'r') as file: # 'r' stands for read mode, you can also use 'w' for write mode, 'a' for append mode, 'x' for execute mode
    content = file.read()
    print(content)
```

```python
# Now, let's demonstrate writing to the file, appending to it, and reading it again

with open('test.txt', 'a+') as file: # 'a+' stands for append and read mode, you can also use 'w+' for write and read mode, 'r+' for read and write mode
    file.write('\nThis is a new line added to the file.')
    file.seek(0)  # Move the cursor to the beginning of the file to read its content
    updated_content = file.read()
    print(updated_content)
```

```python
# using traditional open and close methods

f = open('test.txt', 'w') # open file in write mode
f.write("Hello________") # write something to the file
f.close() # must close the file after writing to it

f = open('test.txt', 'r') # open file in read mode
data = f.read() # read the content of the file
print(data)
f.close() # close the file after reading
```

```python
# class = methods(function in class) + attributes

class student: # static without constructor : you can create multiple objects from this class but all objects will have the same values for the attributes
 
    # attributes

    name = "Adel"
    age = 20
    GPA = 3.5
    gender = "Male"

    # methods

    def student_info(self): # self -> object itself
        # return f"Welcome into DEPI program."
        print(self) # print address of the object itself
```

```python
# create object from the class and access its attributes and methods

s1 = student() # creating an object from the class
s1.name  # accessing attribute of the class
s1.student_info()  # accessing method of the class
```

```python
print(s1.student_info()) # same print(self) # print address of the object itself
```

```python
# constructor __init__ method : special method that is called when an object is instantiated

class Student: # dynamic with constructor : you can create multiple objects from this class with different values for the attributes
    def __init__(self, name, age, GPA, gender): # constructor method
        
        # initialize attributes

        self.name = name 
        self.age = age
        self.GPA = GPA
        self.gender = gender

    # methods
    
    def student_info(self):
        return f"Name: {self.name}, Age: {self.age}, GPA: {self.GPA}, Gender: {self.gender}"
```

```python
s1 = Student("Adel", 20, 3.5, "Male")
print(s1.student_info()) # = s1.student_info() but with return statement
```

```python
# abstract class and abstract method : to create a blueprint for other classes

from abc import ABC, abstractmethod

class Animal(ABC): # abstract class

    @abstractmethod # decorator to define abstract method : change functionality of method
    def sound(self): # abstract method
        pass

class Dog(Animal): # derived class
    def sound(self): # implement abstract method
        return "Woof"
    
class Cat(Animal): # derived class
    def sound(self): # implement abstract method
        return "Meow"
    
class Bird(Animal): # derived class
    pass # does not implement abstract method, will raise an error
```

```python
# example usage

cat_obj = Cat()
print(cat_obj.sound()) # Output: Meow

dog_obj = Dog()
print(dog_obj.sound()) # Output: Woof

# animal_obj = Animal() # This will raise an error because you cannot instantiate an abstract class (just template)
```

```python
# inheritance : to inherit attributes and methods from a parent class to a child class

class Parent:
    def message(self):
        return "This is the parent class message."

class GrandParent:
    def message(self):
        return "This is the grandparent class message."

class Child(Parent):
    def message(self):
        return "This is the child class message."
    
class GrandChild(GrandParent, Child):
    pass

# example usage
parent_obj = Parent()
print(parent_obj.message()) # Output: This is the parent class message.

child_obj = Child()
print(child_obj.message()) # Output: This is the child class message.

grandchild_obj = GrandChild()
print(grandchild_obj.message()) # Output: This is the child class message.
```
