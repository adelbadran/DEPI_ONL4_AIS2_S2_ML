## List (Cont...)
---

```python
# nested_list

list_one = [1, 2, 3, 4, [ 5, 6, 7], 8, 9]

print("Original List:", list_one) # Printing the original list
print("Length of Original List:", len(list_one)) # Length of the list
print(list_one[4][1]) # Accessing the second element of the nested list
```

```python
# functions

list_one = [1, 2, 3, 4, 5]

list_one.append(6) # Appending an element to the list
print("List after append:", list_one)
list_one.insert(0, 2) # Inserting an element at index 2
print("List after insert:", list_one)
list_one.extend([7, 8, 9]) # Extending the list with another list
print("List after extend:", list_one)
list_one.remove(3) # Removing an element from the list
print("List after remove:", list_one)
list_one.pop() # Popping the last element from the list
print("List after pop:", list_one)
del list_one[1] # Deleting the element at index 1
print("List after deleting element at index 1:", list_one)
index = list_one.index(4) # Finding the index of an element notr: returns first occurrence, raises error if not found
print("Index of element 4:", index)
list_one.reverse() # Reversing the list
print("Reversed List:", list_one)
list_one.sort(reverse=True) # Sorting the list in descending order note: default -> reverse=False for ascending
print("Sorted List in descending order:", list_one)
```

```python
# buffer (pop)

buffer = list_one.pop() # Popping the last element and storing it in buffer

print("Buffer after popping from list_one:", buffer)
print("list_one after popping an element:", list_one)
```

```python
# delete

del list_one # Deleting the entire list

# print(list_one) # This will raise an error since list_one is deleted
```

```python
# Clearing the list

list_one.clear() # Clearing all elements from the list

print("List after clearing all elements:", list_one)

# count

list_one = [1, 2, 3, 4, 2, 2, 5]
count_of_twos = list_one.count(2) # Counting occurrences of 2 in the list

print("Count of 2 in list_one:", count_of_twos)
```

```python
# concatination

list_one = [1, 2, 3, 4, 5]
list_two = [10, 11, 12]

combined_list = list_one ,list_two # Concatenating two lists note : use + operator for merging into single list

print("Combined List:", combined_list)
```

```python
# slicing

sliced_list = list_one[1:4] # Slicing the list from index 1 to 3

print("Sliced List (index 1 to 3):", sliced_list)
```

## Set (Cont...)
---

```python
# cannot retrieve index because sets are unordered

set_one = {10, 20, 30, 40, 50}
# item = set_one[2] # This will raise an error since sets do not support indexing
# index = set_one.index(30) # This will raise an error since sets do not support indexing

print("Set elements:", set_one)
```

```python
# functions for set

set_one.add(60) # Adding an element to the set
print("Set after adding 60:", set_one)
set_one.remove(20) # Removing an element from the set
print("Set after removing 20:", set_one)
set_one.discard(100) # Discarding an element (no error if not found)
print("Set after discarding 100 (not present):", set_one)
set_one.pop() # Popping an arbitrary element from the set
print("Set after popping an element:", set_one)
set_one.clear() # Clearing all elements from the set
print("Set after clearing all elements:", set_one)
```

```python
# set operations

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
union_set = set_a.union(set_b) # Union of two sets

print("Union of set_a and set_b:", union_set)

intersection_set = set_a.intersection(set_b) # Intersection of two sets

print("Intersection of set_a and set_b:", intersection_set)

difference_set = set_a.difference(set_b) # Difference of two sets

print("Difference of set_a and set_b (set_a - set_b):", difference_set)
```

# Operations
---

```python
# arithmetic operations

a = 15
b = 4

addition = a + b # Addition
print("Addition:", addition)
subtraction = a - b # Subtraction
print("Subtraction:", subtraction)
multiplication = a * b # Multiplication
print("Multiplication:", multiplication)
division = a / b # Division
print("Division:", division)
floor_division = a // b # Floor Division > prints the largest integer less than or equal to the division result
print("Floor Division:", floor_division)
modulus = a % b # Modulus > prints the remainder of the division
print("Modulus:", modulus)
exponentiation = a ** b # Exponentiation > a ^ b
print("Exponentiation:", exponentiation)
```

```python
# comparison operators

x = 10
y = 20

equal = (x == y) # Equal to
print("x == y:", equal)
not_equal = (x != y) # Not equal to
print("x != y:", not_equal)
greater_than = (x > y) # Greater than
print("x > y:", greater_than)
less_than = (x < y) # Less than
print("x < y:", less_than)
greater_equal = (x >= y) # Greater than or equal to
print("x >= y:", greater_equal)
less_equal = (x <= y) # Less than or equal to
print("x <= y:", less_equal)
```

```python
# logical operators

p = True
q = False

and_result = p and q # Logical AND
print("p and q:", and_result)
or_result = p or q # Logical OR
print("p or q:", or_result)
not_result = not p # Logical NOT
print("not p:", not_result)
```

```python
# identity operators

a = [1, 2, 3]
b = a

identity_check = (a is b) # Identity check
print("a is b:", identity_check)
identity_check_not = (a is not b) # Identity check not
print("a is not b:", identity_check_not)
```

```python
# assignment operators

c = 5

c += 3 # Equivalent to c = c + 3
print("c after += 3:", c)
c -= 2 # Equivalent to c = c - 2
print("c after -= 2:", c)
c *= 4 # Equivalent to c = c * 4
print("c after *= 4:", c)
c /= 2 # Equivalent to c = c / 2
print("c after /= 2:", c)
c %= 3 # Equivalent to c = c % 3
print("c after %= 3:", c)
c //= 2 # Equivalent to c = c // 2
print("c after //= 2:", c)
c **= 3 # Equivalent to c = c ** 3
print("c after **= 3:", c)
```

```python
# task: simple calculator

num_1 = float(input("Enter firset number:"))
num_2 = float(input("Enter second number:"))
op = input("Enter operator (+, -, *, /): ")

if op == '+':
    result =  num_1 + num_2
if op == '-':
    result = num_1 - num_2
if op == '*':
    result = num_1 * num_2
if op == '/':
    result = num_1 / num_2

print(f"{num_1} {op} {num_2} = {result}")
```

# IF condition
---

```python
# if else

a = 200
b = 33

if a > b: # condition = True
    print("a is greater than b")
else: # condition = False
    print("a is not greater than b")
```

```python
# if elif else

a = 100
b = 100

if a > b: # condition 1 > true or false
    print("a is greater than b")
elif a == b: # condition 2 > if condition 1 = false
    print("a is equal to b")
else: # condition 3 > false notherwise
    print("a is less than b")
```

```python
# nested if

a = 50
b = 200
if a < b: # outer if
    print("a is less than b")
    if a % 2 == 0: # inner if
        print("a is also even")
    else: # inner else
        print("a is odd")
else: # outer else
    print("a is not less than b")
```

```python
# don't use == if, if

a = 60
b = 20

if a > b: # condition 1 > True
    print("a is greater than b")
if a > 30: # condition 2 > True
    print("a is greater than 30")
```

```python
# if with logical operators

a = 25
b = 15

if (a > 20 and b < 20): # both conditions true
    print("Both conditions are true")
```

```python
# short hand if

a = 5
b = 15

result = "a is greater than b" if a > b else "a is not greater than b"
print(result)
```

```python
# if elif else short hand

result_1 = "a is greater than b" if a > b else "a is equal to b" if a == b else "a is less than b"
# == 
if a > b:
    result_2 = "a is greater than b"
else:
    if a == b:
        result_2 = "a is equal to b"
    else:
        result_2 = "a is less than b"
        
print(result_1)
print(result_2)
```

# Loop
----

```python
# for loop

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# continue

for number in range(1, 11):
    if number % 2 == 0:
        continue # Skip even numbers
    print("Odd Number:", number)
```

```python
# break

for number in range(1, 11):
    if number == 6:
        break # Stop the loop when number is 6
    print("Number:", number)
```

```python
# while loop

count = 1
while count <= 5:
    print("Count:", count)
    count += 1
```

```python
# break and continue in while loop

count = 1
while count <= 10:
    if count == 6:
        break # Stop the loop when count is 6
    if count % 2 == 0:
        count += 1
        continue # Skip even numbers
    
    print("Odd Count:", count)
    count += 1
```
