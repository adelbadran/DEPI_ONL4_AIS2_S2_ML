## List
----

```python
# one variable = multiple values (multiple assignment)
```

```python
# Data order, changeable, indexed, allows duplicate values, heterogeneous elements, defined by square brackets []

my_list = [1, 2, 3, 4, 5, 'a', True, "adel", 3] # like array in other programming languages

print(type(my_list)) # list type

my_list[0] = 10 # modifying element at index 0

print(my_list) # accessing elements
print(my_list[2:5])  # slicing from index 2 to 4, index 5 is excluded
print(my_list[-1]) # last element
print(my_list[-4:-1]) # slicing from the end, index -1 is excluded note: [-1 : -4] will give empty list
print(len(my_list)) # length of the list
```

```python
name = "adel" # Object must be iterable

list_name = list(name) # converting string to list
print(list_name)
```

## Tuple
----

```python
# Data order, unchangeable, indexed, allows duplicate values, heterogeneous elements, defined by parentheses ()

my_tuple = (1, 2, 3, 4, 5, 'a', True, "adel", 3) # like list but immutable

print(type(my_tuple)) # tuple type
print(my_tuple) # accessing elements
print(my_tuple[2:5])  # slicing from index 2 to 4, index 5 is excluded
print(my_tuple[-1]) # last element
print(my_tuple[-4:-1]) # slicing from the end, index -1 is excluded note: [-1 : -4] will give empty tuple
print(len(my_tuple)) # length of the tuple
```

```python
name = "adel" # Object must be iterable

tuple_name = tuple(name) # converting string to tuple
print(tuple_name)
```

## Set
----

```python
# Data unordered, unchangeable, no duplicate values, defined by curly braces {}

my_set = {1, 2, 3, 4, 5, 'a', True, "adel"} # like list but unordered and no duplicates

print(type(my_set)) # set type
print(my_set) # accessing elements (note: order may vary)
print(len(my_set)) # length of the set
# Note: Sets do not support indexing or slicing
```

```python
name = "Adel Tamer" # Object must be iterable

set_name = set(name) # converting string to set
print(set_name) # rerun to see different order -> not different each time
```

# String
----

```python
# Accessing characters in a string

name = "adel" # [0,1,2,3], [-4,-3,-2,-1]

name[0] # index starts at 0
```

```python
# string slicing

name[1:3] # index 3 is excluded
```

```python
# indexing from the end
name[-1] # last character
```

```python
# slicing from the end

name[-3:-1] # index -1 is excluded
```

```python
# Functions to manipulate strings

sample_str = "hello World!"

print(sample_str.upper()) # to upper case
print(sample_str.lower()) # to lower case
print(sample_str.replace('o', '.')) # replace 'o' with '.'
print(sample_str.capitalize()) # capitalize first letter
print(sample_str.title()) # title case
print(sample_str.swapcase()) # swap case
print(sample_str.split('o')) # split by 'o'
print(sample_str.count('o')) # count occurrences of 'o'
print(sample_str.startswith('h')) # check if starts with 'h'
print(sample_str.endswith('!')) # check if ends with '!'
print(sample_str.find('W')) # find index of 'W'
print(len(sample_str)) # length of the string
print(sample_str.index('W')) # index of 'W'
print(sample_str.isalpha()) # check if all characters are alphabetic
print(sample_str.isdigit()) # check if all characters are digits
print(sample_str.strip('!')) # remove '!' from the end
print("  Hello World!  ".strip()) # remove leading and trailing whitespace
print("  Hello World!  ".lstrip()) # remove leading whitespace
print("  Hello World!  ".rstrip()) # remove trailing whitespace
print(sample_str.center(20, '-')) # center the string with '-' padding
print(sample_str.ljust(20, '-')) # left justify the string with '-' padding
print(sample_str.rjust(20, '-')) # right justify the string with '-' padding
print(sample_str.zfill(20)) # pad the string with leading zeros to a total length of 20
```

```python
split_str = "This is a sample string"

print(split_str.split(' ')) # split by 'space'

response = ''.join(split_str)

print(response)
```
