# Task-1
**Suppose you have this email address “Amit_ml@gmail.edu”**
- Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol.
- If it's not a valid email, eturn "Invalid email".
- Extract Username: Extract and return the part of the email before the "@" symbol.
- Extract Domain: Extract and return the domain (the part between "@" and the last ".").
- Check for Domain Ending: Check if the email ends with ".com". If it does, return "Commercial Domain".
- If it ends with ".edu", return "Educational Domain". Otherwise, return "Other Domain".

```python
def task_1(gmail_input="Amit_ml@gmail.edu"):
    if gmail_input.count('@') != 1:
        print("Invalid email.")
        return

    at_index = gmail_input.index('@')

    if '.' not in gmail_input[at_index:]:
        print("Invalid email.")
        return

    print("Valid email.")

    username = gmail_input.split('@')[0]
    print("Username:", username)

    domain = gmail_input.split('@')[1].rsplit('.', 1)[0]
    print("Domain:", domain)

    if gmail_input.endswith(".com"):
        print("Commercial Domain.")
    elif gmail_input.endswith(".edu"):
        print("Educational Domain.")
    else:
        print("Other Domain.")
```

```python
task_1()
```

# Task-2:
Encoded Message: ###!!@mocleW EPGTQ!!!6789
- Steps to Decode:
1. Extract the core part of the message: "mocleW EPGTQ".
2. Reverse the first word: "mocleW" becomes "Welcome".
3. Replace shifted vowels in the second word: o "EPGTQ": No vowels to change.
4. Final decoded message: "Welcom EPGTQ".

```python
def task_2(input_string="###!!@mocleW EPGTQ!!!6789"):
    core = ""
    for input_char in input_string:
        if (input_char.isalpha() or input_char == ' '):
            core += input_char
    first, second = core.split(' ')
    first = first[::-1]
    reversed = first + ' ' + second
    return reversed
task_2()
```

# Task-3:
Encoded Message: &&&**$gnirtS PLIO!!@1234
- Steps to Decode:
1. Extract the core part of the message: "gnirtS PLIO".
2. Reverse the first word: "gnirtS" becomes "String".
3. Replace shifted vowels in the second word:
o
"PLIO": Replace I->E and O->U to get "PLEU".
4. Final decoded message: "String PLEU".

```python
def task_3(input_string="&&&**$gnirtS PLIO!!@1234"):
    core = ""
    for input_char in input_string:
        if (input_char.isalpha() or input_char == ' '):
            core += input_char
    first, second = core.split(' ')
    first = first[::-1]
    second = second.replace('I', 'E').replace('O', 'U')
    reversed = first + ' ' + second
    return reversed
    
task_3()
```
