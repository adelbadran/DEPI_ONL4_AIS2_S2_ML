```python
# user defined function using def with documentation

def user_info(name : str = "name", age : int = 20): # str, int (hint type)
    """
    this function take name and age from user and return it again
    
    :param_1 : take the name of user from input user interface
    type param_1 : str
    
    :param_2 : take the age of user from input user interface
    type param_2 : int

    return : this function will print the name and the age
    rtype : str, int
    """
    return f"my name is {name}, my age is {age}."
```

```python
# Example usage - positional argument function

print(user_info("adel", 20))

# Example usage - keyword argument function
print(user_info(20, "adel")) # incorrect usage - will assign age to name and name to age
print(user_info(age=20, name="adel")) # solution

# Example usage - default argument function
print(user_info())
```

```python
# factorial function using recursion

def factorial(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n * factorial(n-1)

# example usage
result = factorial(6)
print(result)
```
